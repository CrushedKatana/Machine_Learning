```python
from datasets import load_dataset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.metrics import classification_report, accuracy_score
from sklearn.svm import SVC
from skimage.feature import hog
from skimage import exposure
from skimage.color import rgb2gray
from skimage.transform import resize
from skimage.util import random_noise
from skimage.exposure import equalize_hist
import joblib
```

```python
# Load FairFace dataset (version 0.25 or 1.25)
ds = load_dataset("HuggingFaceM4/FairFace", "0.25")

print(ds)
```

```python
import cv2

def preprocess_image(img):
    img = np.array(img)

    # ---- 1. Crop (center crop) ----
    h, w, _ = img.shape
    min_dim = min(h, w)
    start_h = (h - min_dim) // 2
    start_w = (w - min_dim) // 2
    img = img[start_h:start_h+min_dim, start_w:start_w+min_dim]

    # ---- 2. Resize ----
    img = resize(img, (128, 128), anti_aliasing=True)

    # ---- 3. Convert to grayscale ----
    img_gray = rgb2gray(img)

    # ---- 4. Histogram equalization (enhancement) ----
    img_eq = equalize_hist(img_gray)

    # ---- 5. Noise removal (Gaussian blur) ----
    img_clean = cv2.GaussianBlur(img_eq, (3, 3), 0)

    return img_clean
```

```python
def extract_hog_features(img):
    features, hog_image = hog(
        img,
        orientations=9,
        pixels_per_cell=(8, 8),
        cells_per_block=(2, 2),
        block_norm='L2-Hys',
        visualize=True
    )
    return features
```

```python
X = []
y = []

for sample in ds["train"]:
    img = preprocess_image(sample["image"])
    feat = extract_hog_features(img)

    X.append(feat)
    y.append(sample["race"])

X = np.array(X)
y = np.array(y)

print("Shape X:", X.shape)
print("Labels:", np.unique(y))
```
